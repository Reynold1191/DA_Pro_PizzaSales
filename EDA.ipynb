```python
from typing import List


def candy(ratings: List[int]) -> int:
        n = len(ratings)
        candies = [1] * n 

        for i in range(1, n):
            if ratings[i] > ratings[i-1]:
                candies[i] = candies[i-1] + 1

        for i in range(n-2, -1, -1):
            if ratings[i] > ratings[i+1]:
                candies[i] = max(candies[i], candies[i+1] + 1)
        print(candies)
        return sum(candies)

candy([1,0,2])
```

```python
df = pd.read_excel('data_pizza.xlsx')
ingredients_one_hot = df['ingredients'].str.get_dummies(', ')

ingredients_one_hot
ingredient_counts = ingredients_one_hot.sum().sort_values(ascending=False)

```

```python
top_10_ingredients = ingredient_counts.head(10)
top_10_ingredients
```
